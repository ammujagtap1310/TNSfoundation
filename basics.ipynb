```python
# Arguments

def fancy_power(x, p=2, debug=False):
    """
    Here is a fancy version of power that computes the square of the argument or other powers if p is set
    """
    if debug:
        print( "\"fancy_power\" is called with x =", x, " and p =", p)
    return(x**p)
```

```python
print(fancy_power(5))
print(fancy_power(5,p=3))
```

```python
res = fancy_power(p=8,x=2,debug=True)
print(res)
```

```python
#Classes

class Point(object):
    """
    Class of a point in the 2D plane.
    """
    def __init__(self, x=0.0, y=0.0):
        """
        Creation of a new point at position (x, y).
        """
        self.x = x
        self.y = y
        
    def translate(self, dx, dy):
        """
        Translate the point by (dx , dy).
        """
        # self.x = self.x + dx
        self.x += dx
        self.y += dy
        
    def __str__(self):
        # Called when you print() the object or convert it to a string
        return("Point: ({:.2f}, {:.2f})".format(self.x, self.y))
```

```python
p1 = Point()
print(p1)

p1.translate(3,2)
print(p1)

p2 = Point(1.2,3)
print(p2)
```

```python
# Program to display the Fibonacci sequence up to n-th term

nterms = int(input("How many terms? "))

# first two terms
n1, n2 = 0, 1
count = 0

# check if the number of terms is valid
if nterms <= 0:
   print("Please enter a positive integer")
# if there is only one term, return n1
elif nterms == 1:
   print("Fibonacci sequence upto",nterms,":")
   print(n1)
# generate fibonacci sequence
else:
   print("Fibonacci sequence:")
   while count < nterms:
       print(n1)
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
```
